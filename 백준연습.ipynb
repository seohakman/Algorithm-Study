```python
# 2675
T = int(input())
A = []
B = []

for i in range(T):
  R, S = input().split()
  R = int(R)
  A.append(R)
  B.append(S)

for i in range(len(A)):
  for j in list(B[i]):
    print(j*A[i], end = '')
  print('\n')
```
